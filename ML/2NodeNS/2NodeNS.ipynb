```python
import numpy as np
import matplotlib.pyplot as plt
import random
import pandas as pd
```

```python
df=pd.read_csv(r'D:\\Codes\\HelloWorld\ML\\2NodeNS\\SimpleSample.csv',header=None,sep=',')
datas=df.to_numpy()[1:]
```

```python
def judgeStep(num):
    return 1 if num>0 else 0

learntimes=1000
w1=w2=10
x1=x2=0
c=0
for i in range(learntimes):
    index = random.randint(0,21)
    x1 = datas[index][0]
    x2 = datas[index][1]
    ExpectedAns = datas[index][2]
    Ans = judgeStep(x1*w1+x2*w2+c)
    DeltaAns = Ans - ExpectedAns

    c = c - DeltaAns
    w1 = w1 - DeltaAns * x1 * 0.1
    w2 = w2 - DeltaAns * x2 * 0.1

print(w1,w2,c)
```

```python
def StepFunc(num):
    return 1 if num>0 else 0
learntimes=1000

inputs=np.asarray([0,0])
W=np.asarray([10,10])
c=0
for i in range(learntimes):
    index = random.randint(0,21)
    inputs=datas[index][0:2]
    ExpectedAns = datas[index][2]
    Ans = judgeStep(np.dot(inputs,W)+c)
    DeltaAns = Ans - ExpectedAns

    c = c - DeltaAns
    W = W-DeltaAns*inputs * 0.1
print(W,c)
```

```python
def FuncG(x):
    return (x*w1+c)/-w2
def FuncG2(x):
    return (x*W[0]+c)/-W[1]

fig = plt.figure()
ax = fig.add_subplot(111)
X=np.linspace(0,10,256,endpoint=True)

for i in range(len(datas)):
    ax.scatter(datas[i,0],datas[i,1],color="red" if datas[i,2]==1 else "green")
plt.plot(X,FuncG2(X),color="blue")
plt.show()
```
